```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'student_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110, 111, 112, 113, 114, 115, 116, 117, 118, 119, 120],
    'age': [15, 16, 45, 17, np.nan, 16, 18, 19, 20, 21, 25, 22, 16, np.nan, 17, 50, 18, 19, 16, 25],  # 45, 50 are outliers, np.nan for missing
    'marks': [78, 85, 120, 90, np.nan, 82, 30, 88, 92, 75, 110, 95, 88, 84, 95, 105, 120, np.nan, 80, 60],  # 120, 110, 105 are outliers, np.nan for missing
    'attendance': [85, 92, 150, 88, 90, np.nan, 10, 92, 95, 100, 98, 94, 91, 80, 88, 160, 95, 91, np.nan, 85]  # 150, 160 are outliers, np.nan for missing
}
```

```python
df = pd.DataFrame(data)
```

```python
df.dtypes
```

```python
def plots(column):
    sns.boxplot(df[column])
    plt.title(f"Box Plot of {column}")
    plt.show()
```

```python
for column in ['age','marks','attendance']:
    plots(column)
```

```python
def handle(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR =Q3 - Q1
    lowerbound = Q1 - 1.5*IQR
    upperbound = Q3 + 1.5*IQR

    df[column] = np.where(df[column]<lowerbound , lowerbound , df[column])
    df[column] = np.where(df[column]>upperbound , upperbound , df[column])
```

```python
for column in ['age','marks','attendance']:
    handle(column)
```

```python
for column in ['age','marks','attendance']:
    plots(column)
```
